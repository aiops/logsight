```python
result = []
for tpl in grouped.itertuples():
    tpl.level.update({"prediction":tpl.prediction})
    tpl.level.update({"count": tpl.count})
    tpl.level.update({"@timestamp":str(tpl.Index)})
    result.append(tpl.level)
```
